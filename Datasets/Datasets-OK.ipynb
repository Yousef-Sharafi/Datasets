```python
from sklearn.datasets import load_iris, load_digits, load_wine, load_breast_cancer, load_diabetes, fetch_olivetti_faces

def load_sklearn_dataset(dataset_index):
    # Define a list of dataset loading functions
    dataset_functions = [
        load_iris, load_digits, load_wine, load_breast_cancer,
        load_diabetes, fetch_olivetti_faces
    ]

    # Check if the dataset index is within the valid range
    if 0 <= dataset_index < len(dataset_functions):
        # Load the dataset using the selected function
        dataset = dataset_functions[dataset_index]()
        return dataset
    else:
        return None  # Return None for an invalid dataset index

# Example: Load dataset
dataset = load_sklearn_dataset(5)
if dataset is not None:
    print("Loaded dataset:", dataset.DESCR)
else:
    print("Invalid dataset index")

print("********************************************************")
print("********************************************************")
print("********************************************************")

if dataset is not None:
    X = dataset.data  # Features (X)
    y = dataset.target  # Target values (y)
    print("X (features):")
    print(X[:5])  # Print the first 5 rows of features
    print("\ny (target values):")
    print(y[:5])  # Print the first 5 target values
else:
    print("Invalid dataset index")
```
